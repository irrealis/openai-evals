````python
score_re = re.compile(f"({'|'.join(str(n) for n in range(100, -1, -1))})")

def find_score(message):
  for line in reversed(message.splitlines()):
    matches = score_re.findall(line)
    if matches:
      return int(matches[-1]), line
````

Questions:

- How well does the student model perform on the problem set?
- Which problems are hard for student model?
- How consistent are the grading model's evaluations on the problem set?
- Which problems tend to be graded inconsistently?
- Which student answers are hard for the grading model to evaluate consistently?

- Does the grading model grade correctly, compared to human evaluations?

- Can we identify a general rubric that improves grading consistency? Grading correctness?
- Can we modify problem-specific rubrics to improve grading consistency? Grading correctness?

````python
scoring_dict = dict()

ps_dict = scoring_dict.get('problems', dict())
for pn, p in enumerate(evaluation_set_dict['problems']):
  p_dict = ps_dict.get(p['problem_id'], dict())
  ss_dict = p_dict.get('submissions', dict())
  for sn, s in enumerate(evaluation_set_dict['problems'][pn]['submissions']):
    s_dict = ss_dict.get(s['submission_id'], dict())
    es_dict = s_dict.get('evaluations', dict())

    for en, e in enumerate(evaluation_set_dict['problems'][pn]['submissions'][sn]['evaluations']):
      # Extract score from each grader's evaluation of the submission.
      # Scores are 0-100. Higher is better.
      e_dict = es_dict.get(e['evaluation_id'], dict())
      e_dict['evaluation_id'] = e['evaluation_id']
      e_dict['evaluation_score'] = find_score(e['message'])[0]
      # Save the scoring.
      es_dict[e['evaluation_id']] = e_dict

    # Compute submission score as mean of evaluation scores for this submission.
    # Compute submission grader score as standard deviation of evaluation scores.
    #
    # Submission scores are 0-100. Higher is better.
    # Grader scores measure grading consistence. Lower is better.
    e_scores = pd.Series([float(ed['evaluation_score']) for eid, ed in es_dict.items()])
    s_dict['submission_id'] = s['submission_id']
    s_dict['submission_score'] = Decimal(e_scores.mean())
    s_dict['submission_grader_score'] = Decimal(e_scores.std())
    # Save the scoring.
    s_dict['evaluations'] = list(es_dict.values())
    ss_dict[s['submission_id']] = s_dict

  # Compute problem score as mean of submission scores for this problem.
  # Compute problem grader score as mean of submission grader scores.
  #
  # problem scores are 0-100. Higher is better.
  # Grader scores measure grading consistence. Lower is better.
  s_means = pd.Series([float(sd['submission_score']) for sid, sd in ss_dict.items()])
  s_stds = pd.Series([float(sd['submission_grader_score']) for sid, sd in ss_dict.items()])
  p_dict['problem_id'] = p['problem_id']
  p_dict['problem_score'] = Decimal(s_means.mean())
  p_dict['problem_grader_score'] = Decimal(s_stds.mean())
  # Save the scoring.
  p_dict['submissions'] = list(ss_dict.values())
  ps_dict[p['problem_id']] = p_dict

# Compute problem-set score as mean of problem scores.
# Compute problem-set grader score as mean of problem grader scores.
#
# problem scores are 0-100. Higher is better.
# Grader scores measure grading consistence. Lower is better.
p_means = pd.Series([float(pd['problem_score']) for pid, pd in ps_dict.items()])
p_stds = pd.Series([float(pd['problem_grader_score']) for pid, pd in ps_dict.items()])
scoring_dict['problem_set_score'] = Decimal(p_means.mean())
scoring_dict['problem_set_grader_score'] = Decimal(p_stds.mean())
# Save the scoring.
scoring_dict['problems'] = list(ps_dict.values())
````
